```python
import torch
import torch.nn as nn
from torchinfo import summary
```

**Sequential/ Model Class**

    - First layer ( n_features, output=3)
    - 2nd layer ( 3 inputs and 1 output )
    - Means a perceptron 

```python
class Model(nn.Module):

    def __init__(self, num_features):

        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.ReLU(),
            nn.Linear(3, 1),
            nn.Sigmoid()
        )

    def forward(self, features):
        out = self.network(features)
        return out
```

Create a dataset 
    - 10 rows and 5 cols
    
Then create Model class instance with shape of dataset

```python
# create dataset
features = torch.rand(10, 5)
print(features)

# create model
model = Model(features.shape[1])
model
```

```python
# see the model summary
summary(model)
```

```python
# call model for forward
model(features)
```

```python
# show model weights
model.state_dict()
```
