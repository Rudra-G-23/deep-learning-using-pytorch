```python
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

import warnings
warnings.filterwarnings('ignore')
```

# Dataset

```python
path = "https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
df.head()
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 1:],
    df.iloc[:, 0],
    test_size=0.2
)
```

# Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

# Label Encoding

```python
encoder = LabelEncoder()
# Convert both to a string to ensure consistency
y_train = encoder.fit_transform(y_train.astype(str))
y_test = encoder.transform(y_test.astype(str))
```

# Numpy arrays to PyTorch tnesors

```python
X_train_tensor = torch.from_numpy(X_train).float()
X_test_tensor = torch.from_numpy(X_test).float()

y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

# Dataset

```python
class CustomDataset(Dataset):

    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
```

```python
train_dataset = CustomDataset(X_train_tensor, y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)
```

```python
train_dataset[0]
```

# Data loader

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

# Defining the Model (using NN Module)

All steps included the nn model 
-> Using the nn module

- First, we create a class
- Weights, bias 
- Then create the functions 
    -  Forward calculation
    -  Loss find 
    -  Update the Weights


```python
class MySimpleNN(nn.Module):

    def __init__(self, num_features):

        super().__init__()
        # Perceptron design
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):

        # Update the values
        out = self.linear(features)
        out = self.sigmoid(out)

        return out
```

# Imp Parameters

**Loss**

$$\mathcal{L} = - \Big[ y_{\text{target}} \log(y_{\text{pred}}) + (1 - y_{\text{target}}) \log(1 - y_{\text{pred}}) \Big]$$

```python
LR = 0.1
EPOCHS = 25

# No need for manually finding the loss function
loss_function = nn.BCELoss()
```

# Training Pipeline

```python
#  Inside the training pipeline
# In a loop

# define a loop
    # forward pass
    # loss calculate
    # clear gradients
    # backwrad pass
    # parameters update
    # print the loss in each epoch
```

```python
# create model
model = MySimpleNN(X_train_tensor.shape[1])

# define the optimiser 
optimizer = torch.optim.SGD(model.parameters(), lr=LR)

# define a loop
for epoch in range(EPOCHS):
    
    # forward pass
    y_pred = model(X_train_tensor)
    
    # loss calculate
    loss = loss_function(y_pred, y_train_tensor.view(-1, 1))
    
    # clear gradients
    optimizer.zero_grad()
    
    # backwrad pass
    loss.backward()
    
    # parameters update
    optimizer.step()
    
    # print the loss in each epoch
    print(f"Epoch: {epoch + 1} | Loss: {loss.item()} ")
```

# Model bias and weights

```python
model.linear.bias
```

# Evaluation

```python
# set the model to evaluation mode
model.eval()
```

## Train accuracy 

```python
# Model evaluation using test_loader
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in train_loader:

        # Forward pass
        y_pred = model(batch_features)
        y_pred = (y_pred > 0.5).float()

        # calculate the accuracy
        batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f"Accurac: {overall_accuracy:.4f} ")
```

## Test accuracy 

```python
# Model evaluation using test_loader
accuracy_list = []

with torch.no_grad():
    for batch_features, batch_labels in test_loader:

        # Forward pass
        y_pred = model(batch_features)
        y_pred = (y_pred > 0.5).float()

        # calculate the accuracy
        batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f"Accurac: {overall_accuracy:.4f} ")
```
